### Group by  and Aggrigating

```python
import pandas as pd
df7 = pd.read_csv("/content/Flavors.csv")
df7
```

```python
group_by_frame = df7.groupby('Base Flavor')
type(group_by_frame)
```

```python
group_by_frame.mean(numeric_only = True)
```

```python
group_by_frame.count()
```

```python
group_by_frame.sum()
```

```python
df7.groupby(['Base Flavor', 'Liked']).agg(
    {
        'Flavor Rating': ['mean', 'max', 'count', 'sum'],
        'Texture Rating' : ['mean', 'max', 'count', 'sum']
        })
```

### Data Merge, Concatinate, and Join

```python
import pandas as pd
df1 = pd.read_csv("/content/LOTR.csv")
df2 = pd.read_csv("/content/LOTR 2.csv")
```

```python
df1
```

```python
df2
```

```python
df1.merge(df2, how='inner', on=['FellowshipID', 'FirstName'])
```

```python
df1.merge(df2, how='outer')
```

```python
df1.merge(df2, how='left', on=['Skills'])
```

```python
df1.merge(df2, how='right')
```

```python
# concatinate df1 and df2
pd.concat([df1, df2])
```
